```python
import numpy as np
import pickle 
import pandas as pd
stock = pickle.load(open("stock.pkl",'rb'))
```

```python
time_index = stock['0015']['date']
coins = ['3231', '2610', '2382', '2356', '2618', '2363', '3481', '2891', '2888']
stock_cols = ['date', 'high', 'low', 'open', 'close', 'vol']
```

```python
a = []
for i in coins:
    b = []
    for j in time_index:
        c = []
        for k in range(len(stock_cols)):
            if k == 0:
                c.append(str(stock[i].loc[j,stock_cols[k]]))
            else:
                c.append(stock[i].loc[j,stock_cols[k]])
        c.append(stock[i].loc[j,'close'])
        b.append(np.array(c))
    a.append(np.array(b))
a = np.array(a)
```

```python
np.save('./stock_t.npy', a)
```
